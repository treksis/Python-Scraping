```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://www.marketwatch.com/')
soup = BeautifulSoup(res.content,'html.parser')
data = soup.select('body > section > div.region.region--full.masthead-elements > div > div.element.element--markets.desktop > div > div.markets__table > table > tbody > tr')
data
```

```python
for item in data:
    print(item.find('td', 'table__cell symbol').get_text().strip())
    print(item.find('td', 'table__cell price').get_text().strip())
    print(item.find('td', 'table__cell change').get_text().strip())
    print("\n")
```

```python
for item in data:
    print(item.find('td', 'table__cell symbol').get_text().split('\n'))
    print(item.find('td', 'table__cell price').get_text().split('\n'))
    print(item.find('td', 'table__cell change').get_text().split('\n'))
```

```python
for item in data:
    print(
        item.find('td', 'table__cell symbol').get_text().split('\n')[0],
        item.find('td', 'table__cell price').get_text().split('\n')[1],
        item.find('td', 'table__cell change').get_text().split('\n')[1]
    )
```
