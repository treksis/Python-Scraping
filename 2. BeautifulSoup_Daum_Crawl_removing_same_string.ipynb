```python
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://www.daum.net/')
soup = BeautifulSoup(res.content, 'html.parser')

data = soup.findAll('a','link_issue')
for item in data:
    print(item.get_text())
```

```python
data1 = soup.findAll('a', attrs={'class':'link_issue', 'tabindex':'-1'})
for item in data1:
    print(item.get_text())
```

```python
res = requests.get('https://www.naver.com/')
soup = BeautifulSoup(res.content, 'html.parser')

data2 = soup.findAll('span', 'ah_k')

for item in data2:
    print(item.get_text())
```

```python
res = requests.get('https://www.naver.com/')
soup = BeautifulSoup(res.content, 'html.parser')

data2 = soup.find('div', 'ah_roll_area PM_CL_realtimeKeyword_rolling')
data2
```

```python
items = data2.find_all('span', 'ah_k')
items
```

```python
for item in items:
    print(item.get_text())
```

```python
items1 = data2.find_all('li', 'ah_item')
items1
```

```python
for item in items1:
    print(item.get_text())
```

```python
for item in items1:
    print(item.get_text().strip())
```

```python
for item in items1:
    print(item.get_text().split())
```

```python
for item in items1:
    print(item.get_text().split()[0])
```

```python
for item in items1:
    print(item.get_text().split()[0], "위: ")
```

```python
for item in items1:
    print(item.get_text().split()[0], "위:", item.get_text().split()[1])
```

```python
for item in items1:
    print(item.get_text().split('\n'))
```
